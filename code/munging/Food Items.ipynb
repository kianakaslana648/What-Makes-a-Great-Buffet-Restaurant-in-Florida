```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
rest = pd.read_csv("../buffet_restaurants.csv")
reviews = pd.read_csv("../buffet_reviews.csv")
reviews = reviews.rename(columns={"stars":"review_stars"})
data = rest.merge(reviews, on="business_id")
aspect = pd.read_csv("../buffet_review_aspect.csv")
data = data.merge(aspect, on=["business_id","review_id"])
data["aspect"] = data["aspect"].str.lower()
data["description"] = data["description"].str.lower()
data.loc[data["description"]=="inferior", "polarity"] = -0.5
data = data[data["review_count"] > 6]
data["aspect"] = data["aspect"].astype(str)
data['aspect'] = data.aspect.apply(lambda x: lemmatizer.lemmatize(x))
data.head(3)
```

```python
foods = pd.read_csv("../Dish.csv")
foods = foods[["name"]]
foods = foods.rename(columns={"name": "aspect"})
foods["aspect"] = foods["aspect"].str.lower()
data_food = data.merge(foods, on="aspect")
exclude = ["restaurant", "education", "food", "dishes",
          "choice", 20, "20", "cuisine", "plaza", "-",
          "1", "10", "100", "14", "2", "40", "50", "60",
          "90", "a", "american", "b", "c", "cool", "i.", "in",
          "y", "source", "dinner", "menu", "entrees", "portion",
          "meals", "special", "cups", "fix", "this", "tips", "plate",
          "entree", "lunch", "star", "combination", "selections", "cup", 
           "straw", "fave", "sides", "nature", "baby", "draw", "minute","reception",
          "red", "specialties", "pan", "fresh", "pains", "ketchup", "flies",
          "condiments", "ginger", "ice", "pop", "hollywood", "french", "appetizer",
          "sauces", "side", "pot", "cafe", "dinners", "house", "poop", "kisses",
          "dates", "island", "oder", "collection", "millionaire", "basket",
          "brains", "knife", "tab", "pain", "glass", "exchange", "fine", "the",
          "table", "power", "luxury", "list", "leg", "pepsi", "two", "blue",
          "bomb", "bottle", "bud", "public", "white", "supper", "plain", "program"]
data_food = data_food[~data_food["aspect"].isin(exclude)]
```

```python
rest_food = data_food[["categories", "name", "aspect", "stars"]].drop_duplicates()
rest_food
```

```python
diff_data = []
for food in list(set(rest_food["aspect"])):
    food_yes = rest_food[rest_food["aspect"] == food]
    food_yes_rest = set(food_yes["name"])
    food_no = rest_food[~rest_food["name"].isin(food_yes_rest)]
    rest_yes = food_yes[["name", "stars"]].drop_duplicates()
    rest_no = food_no[["name", "stars"]].drop_duplicates()
    yes_score = rest_yes["stars"].mean()
    no_score = rest_no["stars"].mean()
    yes_count = rest_yes["stars"].count()
    no_count = rest_no["stars"].count()
    diff_data.append([food, yes_score, no_score, 
                      yes_count, no_count, yes_score - no_score])
diff_df = pd.DataFrame(data=diff_data, columns=["food", "yes","no", 
                            "yes_count", "no_count", "diff"])
diff_df
```

```python
diff_df[diff_df["yes_count"] >= 3].sort_values("diff", ascending=False)[:20]
```
