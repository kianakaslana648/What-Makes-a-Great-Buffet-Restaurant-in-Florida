```python
import numpy as np
import pandas as pd
import re
```

```python
data = pd.read_csv("../restaurant_df.csv")
data.head(3)
```

```python
data['restaurant']= (data.categories.str.contains(pat='Restaurant',
                                                          flags=re.IGNORECASE,
                                                          regex=True))
restaurants = data[(data.restaurant==1) 
                & (data.is_open==1)].reset_index(drop=True)
restaurants
```

```python
bar_df = restaurants.groupby("state", as_index=False).agg({"business_id":
                "count"}).sort_values("business_id")
bar_df.to_csv("../rest_by_state.csv", index=False)
```

```python
business_cats=' '.join(restaurants['categories'])

cats=pd.DataFrame(business_cats.split(','),columns=['category'])
x=cats.category.value_counts()[:26]
x
```
