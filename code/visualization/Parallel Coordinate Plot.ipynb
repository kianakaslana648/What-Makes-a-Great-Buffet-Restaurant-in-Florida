```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("plot5_finalset.csv", index_col=0)
df.head(3)
```

```python
def price_func(price):
    if price == 1:
        new="$"
    elif price == 2:
        new="$$"
    elif price == 3:
        new="$$$"
    elif price == 4:
        new="$$$$"
    elif price == 5:
        new="$$$$$"
    else:
        new=0
    return new
df['price'] = df.apply(lambda x: price_func(x['price']), axis=1)
df = df[["business_id", "name", "food_score", "service_score","price"]]
df = df.rename(columns={"star_binning": "star"})
df = df[["business_id", "price"]]
df.head(3)
```

```python
scores = pd.read_csv("scores_all.csv")
scores = scores.merge(df, on="business_id")
scores = scores.fillna(scores.median())
scores.head(3)
```

```python
scores.to_csv("parallel_data.csv", index=False)
```
